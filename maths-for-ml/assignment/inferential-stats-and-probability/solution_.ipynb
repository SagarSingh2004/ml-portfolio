Q1. A company claims that their new battery lasts longer than their previous model, which has a mean lifespan of 100 hours in previous model. A sample of 25 new batteries is tested, yielding a mean lifespan of 105 hours with a standard deviation of 10 hours. At a 5% significance level, can we conclude that the new battery lasts longer?

Formulate the null and alternative hypotheses, choose an appropriate test, and conduct the test to reach a conclusion.

Answer :

<center>Critical Value Method</center>

Step 1: Define null and alternative hypothesis

Step 2: Choose the test

```python
#Given
Sample_mean = 105
Sample_standard_deviation = 10
n = 25
μo = 100
significance_level = 0.05

#Since the population standard deviation is unknown and the sample size is less than 30, So we go for one sample t-test
```

Step 3: Calculate the test_statistic value

```python
#Formuls
t_stats = (Sample_mean - μo) / (Sample_standard_deviation / np.sqrt(n))
t_stats
```

Step 4: Determine the critical value

```python
Dof = n - 1
t_critical = 1.711
print("Dof :",Dof)
print("significance_level :",significance_level)
print(f"From table t_critical : {t_critical}")
```

Step 5: Decision rule

```python
if t_stats > t_critical:
    print("Rejecting the Ho")
else:
    print("Accepting the Ho")
```

Q2. A researcher wants to estimate the average time spent on social media by high school students. A sample of 50 students is selected, and the mean time spent is found to be 3.5 hours with a standard deviation of 0.8 hours. Construct a 95% confidence interval for the true mean time spent on social media by all high school students.

Calculate the confidence interval and interpret the results.


Answer :

Step 1: Identify the data

```python
#Given
n = 50
sample_standard_deviation = 0.8
sample_mean = 3.5
confidence_level = 0.95
significance_level = 1 - confidence_level
#Since the population standard deviation is unknown so we go for t distribution
```

Step 2: Calculate degree of freedom

```python
dof = n - 1
dof
```

Step 3: find t_critical value

```python
t_critical = 2.009
print(f"For {significance_level:.2f} significance level and {dof} dof, the t_critical value is {t_critical}")
```

Step 4: Calculate the Upper and lower CI

```python
#Formula
Upper_CI = sample_mean + t_critical*(sample_standard_deviation / np.sqrt(n))
Lower_CI = sample_mean - t_critical*(sample_standard_deviation / np.sqrt(n))
print(f"CI is {Lower_CI:.2f} to {Upper_CI:.2f}")
```

Step 5: Interpretation

Q3. A survey is conducted to examine whether there is an association between gender (male/female) and preference for a new product (like/dislike). The data is summarized in the table below. Conduct a Chi-square test of independence to determine if there is a significant association between gender and product preference.

State the null and alternative hypotheses, calculate the Chi-square statistic, and interpret the results.


| |Like|Dislike|
|---|---|---|
|Male|20|30|
|Female|25|25|

Answer:

<center>P Value Method</center>

Step 1: Define the null and alternative hypothesis

Step 2: Observed data

```python
data = np.array([[20,30],
                 [25,25]])
data
```

Step 3: Performing ChiSquare test of independence

```python
chi2_stats, p_value, dof, expected = stats.chi2_contingency(data)
print("Chi-square statistic:", chi2_stats)
print("p-value:", p_value)
print("Degrees of freedom:", dof)
print("Expected frequencies:\n", expected)
```

Step 4: Checking ChiSquare is reliable or not

```python
if (expected<5).any():
    print("ChiSquare is unreliable here")
```

Step 4: Conclusion

```python
alpha = 0.05
if p_value<alpha:
    print("Rejecting the Ho : There is significant association between gender and preference for a new product")
else:
    print("Fail to reject Ho : There is no significant association between gender and preference for a new product")
```

Q4. A group of students takes a math test before and after attending a 4-week tutoring session. The test scores before and after the session are recorded. Conduct a appropriate test to determine if the tutoring session has significantly improved the students' test scores.

Formulate the null and alternative hypotheses, choose an appropriate test, and conduct the test to reach a conclusion.


|Student|Before|After|
|---|---|---|
|1|60|75|
|2|72|80|
|3|63|78|
|4|80|82|
|5|69|76|

Answer :

Step 1: Define the null and alternative hypothesis

Step 2: Observed Data

```python
Before = [60,72,63,80,69]
After = [75,80,78,82,76]
```

Step 3: Paired t-test

```python
t_stats, p_two_tailed = stats.ttest_rel(After, Before)
print(t_stats)
print(p_two_tailed)
```

```python
# Since this is a right tailed test
p_one_tailed = p_two_tailed/2 if t_stats>0 else 1-( p_two_tailed/2 )
p_one_tailed
```

Step 4: Conclusion

```python
alpha = 0.05
if p_one_tailed < alpha:
    print("Rejecting the Ho i.e. There is a significant improvement in scores after tutoring session")
else:
    print("Fail to reject the Ho i.e. There is no significant improvement in scores after tutoring session")
```

Q5. A botanist is studying the effect of three different fertilizers on the growth of a particular plant species. The plants are divided into three groups, each receiving a different type of fertilizer (Fertilizer A, Fertilizer B, Fertilizer C). After 8 weeks, the height of the plants in each group is measured in centimeters.

Perform a ANOVA to determine if there is a significant difference in the average plant height between the three fertilizers. Report the F-statistic, p-value, and your conclusion.


|Fertilizer|Fertilizer A|Fertilizer B|Fertilizer C|
|---|---|---|---|
|Plant 1|24|30|22|
|Plant 2|26|32|20|
|Plant 3|28|29|23|
|Plant 4|22|31|21|
|Plant 5|25|33|24|

Answer:

<center>P value method</center>

Step 1 : Define the null and alternative hypothesis

```python
Null Hypothesis (Ho) : All fertilizers results in the same average plant height
Alternative Hypothesis (H1) : At least one fertilizer produces a different average plant height
```

Step 2: Observed data

```python
fertilizer_A = [24, 26, 28, 22, 25]
fertilizer_B = [30, 32, 29, 31, 33]
fertilizer_C = [22, 20, 23, 21, 24]
```

Step 3: Perform ANOVA

```python
f_stats, p_value = stats.f_oneway(fertilizer_A, fertilizer_B, fertilizer_C)
print(f_stats)
print(p_value)
```

Step 4: Conclusion

```python
alpha = 0.05
if p_value < alpha:
    print("Rejectecting the Ho: At least one fertilizer has a different average height")
else:
    print("Fail to reject Ho: All fertilizers results in the same average plant height")
```

Q7. Explain Bayes theorem in detail with some example.

Answer :

Bayes’ theorem is a way to update the probability of a hypothesis based on new evidence.

Mathematically:

                                P(A∣B)=P(B∣A)⋅P(A)/P(B)

Where:

P(A∣B) = posterior probability → probability of A given B happened.
P(B∣A) = likelihood → probability of observing B if A is true.
P(A) = prior probability → initial probability of A before seeing evidence.
P(B) = marginal probability → total probability of B happening.

```python
#Given
P_girls = 0.6
P_boys = 0.4
P_glasses_given_girls = 0.1
P_glasses_given_boys = 0.05
```

```python
P_glasses = (P_glasses_given_girls * P_girls) + (P_glasses_given_boys * P_boys)
```

```python
P_girls_given_glasses = (P_glasses_given_girls*P_girls) / P_glasses
P_girls_given_glasses
```
